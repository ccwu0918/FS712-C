
### 遞增與遞減運算子

*   遞增與遞減運算子的成員：



``` C
a++; // 會先執行整個敘述後，再將 a 的值加 1
++a; // 則是先把 a 的值加 1 後，再執行整個敘述
```




image.png

### 下面的程式是使用遞增運算子的範例：

### prog5_5, 遞增運算子「++」

````python
# Define the base name for the C program dynamically
c_program_name = 'prog5_5'
````

````python
%%writefile {c_program_name}.c
/* prog5_5, 遞增運算子「++」 */
#include <stdio.h>
#include <stdlib.h>
int main(void)
{
   int a=3, b=3;

   printf("a=%d",a);
   printf(", a++的傳回值為%d",a++);	/* 計算a++，並印出其傳回值 */
   printf(", a=%d\n",a);

   printf("b=%d",b);
   printf(", ++b的傳回值為%d",++b); 	/* 計算++b，並印出其傳回值 */
   printf(", b=%d\n",b);

   return 0;
}
````

````python
!gcc {c_program_name}.c -o {c_program_name} && ./{c_program_name}
````

````python
output = !./{c_program_name}
print(output)
````

As you can see, the output `['a=3, a++的傳回值為3, a=4', 'b=3, ++b的傳回值為4, b=4']` is now stored in the Python variable `output` as a list of strings.

## 5.4 學習運算式之資料型態的轉換


### 運算式的型態轉換

*   型態轉換發生在運算子左右兩邊的運算元型態不同時
*   自動型態轉換：

   1. 表示範圍較小的型態轉換成表示範圍較大的型態
   2. 例如： int 和 float 相加，int 會被轉成 float
   3. char 和 int 相加，char 會被轉成 int

*   C 語言所提供的型態中，表示範圍從大到小依序為 double、float、long、int、short，最後才是char




### 型態轉換的範例：

image.png

### prog5_8, 運算式的型態轉換

````python
# Define the base name for the C program dynamically
c_program_name = 'prog5_8'
````

````python
%%writefile {c_program_name}.c
/* prog5_8, 運算式的型態轉換 */
#include <stdio.h>
#include <stdlib.h>
int main(void)
{
   char ch='a';
   short s=-2;
   int i=3;
   float f=5.3f;
   double d=6.28;
   printf("(ch/s)-(d/f)-(s+i)=%f\n",(ch/s)-(d/f)-(s+i));
   printf("size=%d\n",sizeof((ch/s)-(d/f)-(s+i)));

   return 0;
}
````

````python
!gcc {c_program_name}.c -o {c_program_name} && ./{c_program_name}
````

````python
output = !./{c_program_name}
print(output)
````

### 運算式中，變數型態的轉換過程：



``` python
   char ch='a';
   short s=-2;
   int i=3;
   float f=5.3f;
   double d=6.28;
```

As you can see, the output `['(ch/s)-(d/f)-(s+i)=-50.184906', 'size=8']` is now stored in the Python variable `output` as a list of strings.
