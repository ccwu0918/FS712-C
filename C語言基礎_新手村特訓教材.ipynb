#### 從鍵盤讀取資料

- **&** - 取地址符 (Address-of Operator)

#### 💡 電機系比喻

# 🔀 If-Else 條件判斷

#### 📋 流程圖 (Flowchart)

```
[開始] → [條件判斷?] ←是/否→ [程式碼 A] / [程式碼 B] → [結束]

- 如果條件成立，執行 A

```

### 幻燈片 11：switch-case 多重選擇

# 🔃 switch-case

#### 📋 多重選擇流程圖

```
[開始] → [變數值?] → [case 1] [case 2] [...] → [對應程式碼] → [break] → [結束]
```

#### 核心要點

- 適合處理多個固定值的選擇
- 記得加 break 防止穿透

#### ⚡ 電機系範例

多工器 (Multiplexer) 選擇通道

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

int main() {
    int channel = 2;

    switch (channel) {
        case 1:
            printf("通道 1\n");
            break;
        case 2:
            printf("通道 2\n");
            break;
        default:
            printf("無效通道\n");
    }

    // 輸出: 通道 2
    return 0;
}
````

### 幻燈片 12：while 與 for 迴圈

# 🔄 迴圈 (Loop)

#### while 迴圈

**條件判斷 → 執行 → 回傳判斷**

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

int main() {
    int i = 1;
    while (i <= 5) {
        printf("%d ", i);
        i++;
    }
    // 輸出: 1 2 3 4 5
    return 0;
}
````

#### for 迴圈

**初始 → 判斷 → 遞增 一氣呵成**

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

int main() {
    for (int i = 1; i <= 5; i++) {
        printf("%d ", i);
    }
    // 輸出: 1 2 3 4 5
    return 0;
}
````

### 幻燈片 14：巢狀迴圈

# 🔺 巢狀迴圈

#### 什麼是巢狀迴圈？

一個迴圈裡面包著另一個迴圈，就像俄羅斯套娃一樣！

- 外層迴圈控制「列」
- 內層迴圈控制「行」

- 矩陣運算
- 二維陣列存取

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

int main() {
    int n = 5;

    for (int i = 1; i <= n; i++) {
        for (int j = 1; j <= i; j++) {
            printf("* ");
        }
        printf("\n");
    }
    return 0;
}
````

### 幻燈片 19：全域變數與區域變數

# 🌍 全域 vs 區域變數

#### 全域變數 (Global)

- 宣告在所有函式外部
- 程式任何地方都能存取
- 佔用記憶體直到程式結束
- ⚠️ 謹慎使用，容易造成錯誤

#### 區域變數 (Local)

- 宣告在函式內部
- 只在該函式內有效
- 函式結束後自動釋放
- ✅ 建議優先使用

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

// 全域變數
int globalVar = 100;

void test() {
    // 區域變數
    int localVar = 50;
    printf("%d\n", globalVar);
    printf("%d\n", localVar);
}

int main() {
    test();
    printf("%d\n", globalVar);
    // printf("%d", localVar); 錯誤！
    return 0;
}
````

### 幻燈片 20：成績管理系統案例

# 📋 成績管理系統

用陣列與函式計算 5 位學生平均成績

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

// 計算平均值的函式
float calcAverage(int arr[], int n) {
    int sum = 0;
    float avg;
    for (int i = 0; i < n; i++) {
        sum += arr[i];
    }
    avg = (float)sum / n;
    return avg;
}

// 主程式
int main() {
    int scores[5] = {90, 85, 78, 92, 88};  // 5位學生成績
    float avg = calcAverage(scores, 5);
    printf("平均成績: %.2f\n", avg);
    return 0;
}

// 輸出: 平均成績: 86.60
````

### 幻燈片 23：指標基礎概念

# 📬 指標基礎概念

#### 🏠 信箱比喻

想象一條街道上的信箱：

- **變數** = 信箱裡的內容
- **指標** = 信箱的地址
- **&** = 查詢地址
- ***** = 打開信箱看內容

#### 記憶體對應

```
[0x0010] → x = 10
    ↓
[0x0014] → p = 0x0010
```

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

int main() {
    // 宣告指標變數
    int x = 10;
    int *p;

    // & 取地址符
    p = &x;

    // * 取值符
    printf("%d\n", *p);  // 輸出: 10
    printf("%p\n", p);   // 輸出: 0x0010
    return 0;
}
````

### 幻燈片 24：指標與陣列關係

# 📊 指標與陣列

#### 重要觀念

陣列名稱本身就是一个指標常數，它指向陣列的第一個元素！

#### 對應關係

```
arr[0] 等同於 *(arr + 0)
arr[1] 等同於 *(arr + 1)
arr[2] 等同於 *(arr + 2)
```

#### 💡 電機系比喻

就像記憶體位址的位移運算，指標算術類似位址偏移！

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

int main() {
    int arr[5] = {10, 20, 30, 40, 50};
    int *p = arr;

    // 三種方式存取陣列
    printf("%d\n", arr[0]);      // 輸出: 10
    printf("%d\n", *arr);        // 輸出: 10
    printf("%d\n", *(p + 2));    // 輸出: 30

    // 指標遞增
    p++;
    printf("%d\n", *p);          // 輸出: 20
    return 0;
}
````

### 幻燈片 25：動態記憶體配置

# 💾 動態記憶體配置

#### malloc() - 配置記憶體

```c
void* malloc(size_t size)
```

配置指定大小的記憶體空間

#### free() - 釋放記憶體

```c
void free(void* ptr)
```

釋放已配置的記憶體

#### ⚠️ 重要原則

每一個 malloc() 都必須對應一個 free()！

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

int main() {
    // 配置 5 個整數的空間
    printf("配置 5 個整數的空間!\n");
    int *arr = (int*)malloc(sizeof(int) * 5);

    if (arr == NULL) {
        // 配置失敗處理
        printf("記憶體配置失敗處理!\n");
    }

    // 使用配置記憶體
    printf("使用配置記憶體!\n");
    for (int i = 0; i < 5; i++) {
        arr[i] = (i + 1) * 10;
    }

    // 使用完畢釋放記憶體
    printf("使用完畢釋放記憶體!\n");
    free(arr);
    return 0;
}
````

### 幻燈片 26：記憶體危險地帶

# ⚠️ 記憶體危險地帶

#### 💀 懸空指標 (Dangling Pointer)

指標指向的記憶體已被釋放，但指標仍保留舊位址

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

int main() {
    int *p = (int*)malloc(sizeof(int));
    free(p);
    // p 現在是懸空指標！
    *p = 100;  // 錯誤！
    prinf("%d", p)
    return 0;
}
````

#### 💧 記憶體洩漏 (Memory Leak)

配置的記憶體沒有釋放，導致可用記憶體越來越少

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

void leak() {
    int *p = (int*)malloc(100);
    // 函式結束時沒 free()
    return;
}

// 每次呼叫洩漏 100 bytes

int main() {
    printf("第一次呼叫leak()函數\n");
    leak();
    printf("第二次呼叫leak()函數\n");
    leak();
    return 0;
}
````

### 幻燈片 27：指標交換變數

# 🔄 指標交換變數

#### swap(&a, &b) 執行過程

```
a = 5 → a = 10
b = 10 → b = 5
```

#### 💡 電機系比喻

就像透過指標「指向」交換兩個記憶體位置的值！

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

// 交換函式 (傳指標)
void swap(int *x, int *y) {
    int temp = *x;
    *x = *y;
    *y = temp;
}

// 主程式
int main() {
    int a = 5, b = 10;
    printf("交換前: a=%d, b=%d\n", a, b);
    swap(&a, &b);
    printf("交換後: a=%d, b=%d\n", a, b);
    return 0;
}

// 輸出: 交換前: a=5, b=10
// 輸出: 交換後: a=10, b=5
````

### 幻燈片 29：結構體定義

# 🏗️ 結構體 (struct)

#### 結構體讓你自定義資料型態，將不同類型的變數整合在一起！

#### 💡 就像電機系的元件規格：

- 電阻需要：阻值、容差、功率
- 電容需要：容值、耐壓、類型

#### ⚡ 使用成員運算子 (.):

person.name, person.age

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

// 定義結構體
struct Person {
    char name[50];
    int age;
    char phone[20];
};

// 主程式
int main() {
    // 使用結構體
    struct Person p1;
    strcpy(p1.name, "王小明");
    p1.age = 20;
    strcpy(p1.phone, "0912345678");

    printf("姓名：%s\n", p1.name);
    printf("年齡：%d\n", p1.age);
    printf("電話：%s\n", p1.phone);
    return 0;
}
````

### 幻燈片 30：檔案讀寫

# 📂 檔案讀寫函式

| 函式 | 說明 |
|------|------|
| fopen() | 開啟檔案 |
| fclose() | 關閉檔案 |
| fprintf() | 格式化寫入檔案 |
| fscanf() | 格式化讀取檔案 |

#### ⚡ 檔案模式：

"r" 讀取 | "w" 寫入 | "a" 附加

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

int main() {
    // 寫入檔案
    FILE *fp = fopen("data.txt", "w");
    fprintf(fp, "Name: John\n");
    fprintf(fp, "Age: %d\n", 20);
    fclose(fp);

    // 讀取檔案
    fp = fopen("data.txt", "r");
    char name[50];
    int age;
    fscanf(fp, "Name: %s\n", name);
    fscanf(fp, "Age: %d\n", &age);
    fclose(fp);

    printf("姓名：%s\n", name);
    printf("年齡：%d\n", age);
    return 0;
}
````

### 幻燈片 31：通訊錄管理系統

# 📱 通訊錄管理系統

整合結構體與檔案讀寫

````python
%%cuda
#include <stdio.h>
#include <stdlib.h>

// 定義聯絡人結構體
struct Contact {
    char name[50];
    char phone[20];
    char email[50];
};

// 儲存聯絡人到檔案
void saveContact(struct Contact c) {
    FILE *fp = fopen("contacts.txt", "a");
    fprintf(fp, "%s %s %s\n", c.name, c.phone, c.email);
    fclose(fp);
}

// 主程式
int main() {
    struct Contact p = {"王小明", "0912345678", "john@email.com"};
    saveContact(p);
    printf("已儲存聯絡人！\n");
    return 0;
}
````
