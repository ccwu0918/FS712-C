````python
%%writefile prog1_2.c
/* prog1_2, 第二個程式碼 */
#include <stdio.h>
#include <stdlib.h>
int main(void) {
  char ch;
  printf("Input a character:");
  scanf("%c", &ch);
  printf("ch=%c, The ASCII code is %d\n",ch, ch);
  system("pause");
  return 0;
}
````
