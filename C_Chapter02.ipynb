### prog2_1, 簡單的Ｃ語言

````python
%%writefile prog2_1.c
/* prog2_1, 簡單的Ｃ語言 */
#include <stdio.h>		/* 把stdio.h這個檔案含括進來 */
#include <stdlib.h>		/* 把stdlib.h這個檔案含括進來 */
int main(void)			/* 主函數main()從這兒開始 */
{
   int num;				/* 宣告整數變數num*/
   num=2;		 		/* 把num的值設為2 */
   printf("I have %d cats.\n",num); 	/* 呼叫printf()函數 */
   printf("You have %d cats, too.\n",num); /* 呼叫printf()函數 */
   system("pause");	/* 呼叫dos裡的pause指令，用來暫停程式的執行 */
   return 0;
}
````

## 2.2 解析 C 語言


### prog2_2, 函數的本體與程式區塊

````python
%%writefile prog2_2.c
/* prog2_2, 函數的本體與程式區塊 */
#include <stdio.h>
#include <stdlib.h>
int main(void)
{
   int i=2;	/* 宣告整數i，並設值為2 */
   if(i<5)	/* if區塊由此開始 */
   {
    printf("變數i的值小於5");
    printf("\n");	/* 換行 */
   }				/* if區塊結束 */
   return 0;
}
````

````python
!gcc prog2_2.c -o prog2_2
````

````python
!./prog2_2
````

### prog2_6, 有縮排的程式碼

````python
%%writefile prog2_6.c
/* prog2_6, 有縮排的程式碼  */
#include <stdio.h>
#include <stdlib.h>

int main(void)
{
   int i;
   for(i = 5; i > 0; i--)
   {
      printf("i = %d. Cats are running, ", i);
      printf("dogs are chasing.\n");
   }
   return 0;
}
````
